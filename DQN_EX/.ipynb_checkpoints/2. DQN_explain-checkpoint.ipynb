```python
import sys
sys.path.append('/home/rhino/anaconda3/envs/rl_env/lib/python3.9/site-packages')
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import gym

from collections import namedtuple

# Execute DQN
import random
import torch
from torch import nn
from torch import optim
import torch.nn.functional as F

def printList(name, list):
    print(name + " = ")
    for data in list:
        print(data)
    print()
    
torch.manual_seed(0)
```

```python
namedTuple_ex = namedtuple('nt', ('A', 'B'))
nt_1 = namedTuple_ex('data1', 100)
print("namedTuple = ", nt_1)
```

```python
# data resize
dataList = []
for i in range(10):
    dataList.append((torch.FloatTensor([i*1]), torch.FloatTensor([i*2])))
    
printList("dataList", dataList)


# random
dataSample = random.sample(dataList, 4)
printList("dataSample", dataSample)

# *
print("*dataSample = ", *dataSample)
'''
(tensor([2.]), tensor([4.])) 
(tensor([0.]), tensor([0.])) 
(tensor([5.]), tensor([10.])) 
(tensor([1.]), tensor([2.]))
'''

# zip
print("*zip(*dataSample) = ", *zip(*dataSample))
'''
(tensor([2.]), tensor([0.]), tensor([5.]), tensor([1.])) 
(tensor([4.]), tensor([0.]), tensor([10.]), tensor([2.]))
'''

namedTuple_ex1 = namedtuple('nt1', ('A', 'B'))
# cat
tupleSample = namedTuple_ex1(*zip(*dataSample))
printList("tupleSample", tupleSample)
'''
'A' = (tensor([6.]), tensor([0.]), tensor([5.]), tensor([9.]))
'B' = (tensor([12.]), tensor([0.]), tensor([10.]), tensor([18.]))
'''

print("torch.cat(tupleSample.A) = ", torch.cat(tupleSample.A))
print("torch.cat(tupleSample.B) = ", torch.cat(tupleSample.B))
'''
torch.cat(tupleSample.A) =  tensor([3., 4., 0., 2.])
torch.cat(tupleSample.B) =  tensor([6., 8., 0., 4.])
'''

# gather
sampleA = torch.cat(tupleSample.A)
sampleB = torch.cat(tupleSample.B)
'''
torch.cat(tupleSample.A) =  tensor([6., 2., 8., 4.])
torch.cat(tupleSample.B) =  tensor([12.,  4., 16.,  8.])
'''
print(sampleA.gather(0, torch.tensor([0, 1])))
'''tensor([6., 2.])'''

# cat, shape, reshape
sampleAB = torch.cat([sampleA, sampleB], dim=0)
print(sampleAB)
print(sampleAB.shape)
'''
tensor([ 6.,  2.,  8.,  4., 12.,  4., 16.,  8.])
torch.Size([8])
'''

print(sampleA.shape)
'''torch.Size([4])'''


print(sampleA.reshape(1, 4))
'''tensor([[8., 3., 9., 5.]])'''
sampleA = sampleA.reshape(1, 4)
sampleB = sampleB.reshape(1, 4)
sampleAB = torch.cat([sampleA, sampleB], dim=0)
print(sampleAB)
print(sampleAB.shape)
'''
tensor([[ 5.,  3.,  4.,  0.],
        [10.,  6.,  8.,  0.]])
torch.Size([2, 4])
'''
sampleAB = torch.cat([sampleA, sampleB], dim=1)
print(sampleAB)
print(sampleAB.shape)
'''
tensor([[ 1.,  2.,  6.,  4.,  2.,  4., 12.,  8.]])
torch.Size([1, 8])
'''


# max
sampleAB = torch.cat([sampleA, sampleB], dim=0)
'''
tensor([[ 5.,  8.,  6.,  3.],
        [10., 16., 12.,  6.]])
'''
print(sampleAB.max())
'''tensor(16.)'''
print(sampleAB.max(1))
'''
torch.return_types.max(
values=tensor([ 8., 16.]),
indices=tensor([1, 1]))
'''
print(sampleAB.max(0))
'''
torch.return_types.max(
values=tensor([10., 16., 12.,  6.]),
indices=tensor([1, 1, 1, 1]))
'''


# unsqueeze
print(sampleAB.shape)
'''torch.Size([2, 4])'''
print(sampleAB.unsqueeze(0))
print(sampleAB.unsqueeze(0).shape)
'''
tensor([[[ 6.,  0.,  1.,  2.],
         [12.,  0.,  2.,  4.]]])
torch.Size([1, 2, 4])
'''
print(sampleAB.unsqueeze(1))
print(sampleAB.unsqueeze(1).shape)
'''
tensor([[[ 6.,  0.,  1.,  2.]],

        [[12.,  0.,  2.,  4.]]])
torch.Size([2, 1, 4])
'''

# squeeze
sampleAB_1 = sampleAB.unsqueeze(1)
print(sampleAB_1)
print(sampleAB_1.shape)
'''
tensor([[[ 8.,  4.,  7.,  5.]],

        [[16.,  8., 14., 10.]]])
torch.Size([2, 1, 4])
'''
print(sampleAB_1.squeeze())
print(sampleAB_1.squeeze().shape)
'''
tensor([[ 8.,  4.,  7.,  5.],
        [16.,  8., 14., 10.]])
torch.Size([2, 4])
'''

```
