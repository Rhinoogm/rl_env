```python
import sys
sys.path.append('/home/rhino/anaconda3/envs/rl_env/lib/python3.9/site-packages')
import gym

'''
관찰(observation) (object): 환경에 대한 관찰을 나타내는 객체
보상(reward) (float): 이전의 행동을 통해 얻어지는 보상의 양. 목표는 언제나 보상의 총량을 높이는 것
done(boolean): 환경을 reset 해야할지 나태내는 진리값. ex) done=True일 때 에피소드가 종료
info(dict): 디버깅에 유용한 진단 정보

              anction(->)
Agent                            Environment
        (<-)observation, reward

'''

env = gym.make('CartPole-v1')
# start first observation when executing reset()
observation = env.reset()
# position, speed, angle, turnover
print(observation)

# sampling one of the available actions
action = env.action_space.sample()
print(action)

# step returns the same as observation
# + reward, done, info
step = env.step(action)
print("step = ", step)

for i_episode in range(20):
    observation = env.reset() # First observation
    
    for t in range(100): # For 100 time steps
        env.render()
        print("observation = ", observation)
        action = env.action_space.sample() # Take a random action
        observation, reward, done, info = env.step(action)
        
        if done: # Finish the episode if done
            print('Episode finished after {} timesteps'.format(t+1))
            break
env.close()
        
```
